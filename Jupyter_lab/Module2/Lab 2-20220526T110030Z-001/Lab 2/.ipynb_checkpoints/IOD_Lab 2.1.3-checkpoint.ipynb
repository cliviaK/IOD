The purpose of today's lab is to use simulation to visualize the sampling distribution for the sample mean. The Central Limit Theorem (CLT) tells us that as our sample size gets larger, the sampling distribution of the sample mean converges to a normal distribution. Therefore, when we have a large sample size, we can say that the sampling distribution for the sample mean is approximately normal, regardless of the distribution from which we are sampling.

#### Simulated sampling (sample means)

Now, we'd like to get an idea of what happens when we take multiple random samples of size 5. 

Take 10 sample (size=5) from the entire population. Calculate means for each sample. Now make a histogram of all the sample means.

- Describe the shape of the histogram.
- What is the center of the distribution of sample means?

##### 4. Simulation with `sample_size=5`

Take 10 samples, but with a sample size of 5.

```python
# ANSWER
def take_sample_mean(df, column_name, N=10, sample_size=5):
    '''
    df: dataframe
    column_name: name of column
    N: the number of samples
    sample_size: the size of samples
    '''
    sample_mean_list = []
    for i in range(N):
        df_sample = df.sample(n=sample_size)
        sample_mean = df_sample[column_name].mean()
        sample_mean_list.append(sample_mean) 
        
    return sample_mean_list

def look_sample_mean(population_mean, sample_mean_list):
    '''
    population_mean: mean of entire of population
    sample_mean_list: list of sample mean
    '''
    ax = sns.histplot(sample_mean_list);
    sample_mean = np.mean(sample_mean_list)
   
    ax.axvline(population_mean, color='black', linestyle='solid', lw=1)
    ax.axvline(sample_mean, color='red', linestyle='dashed', lw=1)
    # plot a vertical line for population mean and sample mean
    
    plt.xlabel('Mean of Samples')
    plt.ylabel('Frequency')
    
    plt.show()
```

```python
income_sample_mean_list = take_sample_mean(df= house_hold, column_name = 'Total Household Income', N=10, sample_size=5)
income_sample_mean_list
```

###### 4.A Make a histogram of all the sample means

```python
# ANSWER
look_sample_mean(population_mean= mean_income, sample_mean_list= income_sample_mean_list)
```

###### 8.C As the sample size grew, did your results confirm the CLT?

Yes.

To summarize, the distribution of sample means will be approximately normal as long as the sample size is large enough.

We will depend on the Central Limit Theorem again and again in order to do normal probability calculations when we use sample means to draw conclusions about a population mean. We now know that we can do this even if the population distribution is not normal.

which means we don't need to worry too much about the distribution that the sample came from
we can use the mean's normal distribution to make confidence intervals
do t-tests, where we ask if there is a difference between the means from samples and ANOVa,
where we ask if there is a difference among the means from three or more samples
and pretty much any statistical test that uses the sample mean.

In order for the CLT to work at all, you have to be able to calculate a mean from your sample
