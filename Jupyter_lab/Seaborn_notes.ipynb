```python
crash_df = sns.load_dataset('car_crashes')
crash_df.head()
```

Distribution Plot

```python
sns.distplot(crash_df['not_distracted'])  # kde=False can remove the lines, bins=25
```

Joint Plot  # compare two distributions

```python
sns.jointplot(data= crash_df, x='speeding', y='alcohol')  # kind='reg', kind='kde', kind='hex'
```

```python
sns.jointplot(data= crash_df, x='speeding', y='alcohol',kind='kde')
```

```python
sns.jointplot(data= crash_df, x='speeding', y='alcohol',kind='hex')
```

KDE Plot

```python
sns.kdeplot(crash_df['alcohol'])
```

Pair Plots  # plot the relationships across the entire dataframe numerical values

```python
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
iris_g = sns.PairGrid(iris_copy, hue='species')
```

```python
iris_g = sns.PairGrid(iris_copy, hue='species')
iris_g.map(plt.scatter)
```

```python
iris_g = sns.PairGrid(iris_copy, hue='species')
iris_g.map(plt.scatter)
iris_g.map_diag(plt.hist)
```

```python
iris_g = sns.PairGrid(iris_copy, hue='species')
iris_g.map(plt.scatter)
iris_g.map_diag(plt.hist)
iris_g.map_offdiag(plt.scatter)
```

```python
iris_g = sns.PairGrid(iris_copy, hue='species')
iris_g.map(plt.scatter)
iris_g.map_diag(plt.hist)
iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)  # kde is seaborn plot
```

```python
iris_g = sns.PairGrid(iris_copy,  hue = 'species',
                      x_vars = ['sepal_length', 'sepal_width'],
                      y_vars = ['petal_length', 'petal_width'])
iris_g.map(plt.scatter)
iris_g.add_legend()   # add legend of species color indicators
```

# Facet Grid

```python
tips_fg = sns.FacetGrid(tips_df, col='time', row='smoker')
```

```python
tips_fg = sns.FacetGrid(tips_df, col='time', row='smoker')
tips_fg.map(plt.hist, 'total_bill', bins=8)
```

```python
tips_fg = sns.FacetGrid(tips_df, col='time', row='smoker')
tips_fg.map(plt.scatter, 'total_bill', 'tip')
```

```python
tips_fg = sns.FacetGrid(tips_df, col='time', hue='smoker', height = 4, aspect = 1.3)
# aspect is the width based on the height
tips_fg.map(plt.scatter, 'total_bill', 'tip')
```

```python
tips_fg = sns.FacetGrid(tips_df, col='time', hue='smoker', height = 4,
                        aspect = 1.3,col_order=['Dinner', 'Lunch'], 
                        palette = 'Set1'
                       )
# aspect is the width based on the height
tips_fg.map(plt.scatter, 'total_bill', 'tip')
```

```python
kws = dict(s=50, linewidth=0.5, edgecolor='w')
tips_fg = sns.FacetGrid(tips_df, col='sex', hue='smoker', height=4,
                        aspect=1.3, hue_order =['Yes', 'No'],
                        hue_kws = dict(marker=['^', 'v']))
tips_fg.map(plt.scatter, 'total_bill', 'tip', **kws)
```
